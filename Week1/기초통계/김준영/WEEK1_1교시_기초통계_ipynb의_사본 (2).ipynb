## A1. 답안 작성  
답변:A/B 테스트는 웹/앱의 두 버전( 기존, 변경안)을 사용자에게 무작위로 나눠 노출한 뒤, 성과 지표(CTR, CVR, 체류시간 등)를 비교해서 어느 버전이 더 효과적인지를 데이터로 판단하는 실험 방법이다.


1.   무작위 배정이 제대로 되어야 하며(가능하면 유저 단위로 고정), 실험 중 다른 요소(유입채널, 노출시간, 프로모션 등)는 최대한 동일하게 유지해 교란요인을 줄여야 한다.
2.   실험 전에 가설(H0/H1), 메인 지표, 유의수준을 미리 정하고, 중간 결과를 보고 지표를 바꾸거나 유리할 때만 멈추는 행동(피킹)을 피해야 한다.
3. 충분한 표본 수와 기간을 확보해야 하며, 표본이 작으면 우연 변동으로 잘못된 결론이 나올 수 있다.
4. 초두효과(새로움 효과)를 고려해야 한다. 새로운 UI/문구/혜택은 초반에 신기함이나 낯설음 때문에 반응이 일시적으로 과대/과소평가될 수 있으므로, 일정 기간 이상 실험을 진행하거나 초반·후반을 나눠 확인하는 것이 좋다.
5. 결과 해석 시 통계적 유의성(p-value)뿐 아니라 실제 서비스에 의미 있는 개선인지(효과크기)인 실무적 유의성도 함께 판단해야 한다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   





또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:
*   		p1: **Control(A, 기존안)**에서의 비로그인 유저 기준 주요 지표(여기서는 공모전 카드 클릭률, CTR) 발생 확률
	•	가정: p1 = 0.03 (3.0%)

*       	p2: **Variant(B, 실험안)**에서의 비로그인 유저 기준 공모전 카드 클릭률(CTR) 발생 확률
	•	가정: p2 = 0.04 (4.0%)  (3.0% → 4.0%로 1.0%p 개선 기대)

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.
import math
from scipy.stats import norm

# given
alpha = 0.05
power = 0.80
beta = 1 - power

# 가정한 p1, p2
p1 = 0.03             # Control의 CTR
p2 = 0.04             # Variant의 CTR

# Z values
z_alpha_2 = norm.ppf(1 - alpha/2)  # 양측검정
z_beta = norm.ppf(1 - beta)

# sample size per group
n = ((z_alpha_2 + z_beta)**2 * (p1*(1-p1) + p2*(1-p2))) / ((p1 - p2)**2)

n_per_group = math.ceil(n)
total_n = 2 * n_per_group

print("Z_alpha/2:", z_alpha_2)
print("Z_beta:", z_beta)
print("n (per group):", n_per_group)
print("total n (A+B):", total_n)

#########
```

**3. 지표 및 통계 설계 (Metrics & Statistics)**

> **핵심 측정 지표**
> * **Primary Metric (주지표):** 공모전 카드 클릭률(CTR) = (공모전 카드 클릭 수) / (공모전 리스트 노출 수)  (비로그인 세션 기준)
> * **Secondary Metric (보조지표):**
    - 필터 사용률(Filter Usage Rate) = (관심분야 필터 변경 세션 수) / (홈 진입 세션 수)
    - 추천 시작 전환율(Start Rate) = (“추천 시작하기” 클릭 세션 수) / (홈 진입 세션 수)
> * **Guardrail Metric (방어지표):**
    - 이탈률(Bounce Rate) = (홈에서 추가 행동 없이 종료한 세션) / (홈 진입 세션)
    - 오류율(Error Rate) = (필터/리스트 로딩 실패, 클릭 이벤트 누락 등) / (홈 진입 세션)

| 통계적 설계 항목 | 설정값 | 비고 |
| :--- | :--- | :--- |
| **Significance Level ($\alpha$)** | 5% | 보통 5%(0.05) 설정 |
| **Statistical Power ($1-\beta$)** | 80% | 보통 80% 설정 |

 **4. 위험 관리 및 중단 기준 (Risk & Andon)**

* **Trade-Off :**    
    - 기존의 브랜드 메시지(“누구나 도전을…”) 노출이 줄어들어 브랜딩/감성 전달이 약해질 수 있음
	- 필터 UI가 중앙에 오면서 정보가 많아져 첫 화면이 복잡해 보일 위험
	- 좌측 라디오 삭제로 인해 일부 유저는 기존 위치에 익숙했던 탐색 방식이 바뀌어 혼란을 느낄 수 있음

* **Andon (중단 기준) :**
    1.	주지표(공모전 카드 CTR)가 Control 대비 10% 이상 하락(상대 기준)하고, 같은 추세가 연속 1일 이상 지속되면 즉시 중단/롤백
	    •	예: CTR 3.0% → 2.7% 이하

    2.	기술/운영 이슈 발생 시 중단
	    •	필터/리스트 로딩 실패 등 오류율이 2% 이상으로 상승하거나
	    •	“필터가 안 된다/공모전이 안 뜬다” 등 유저 불만(CS/피드백)이 10건 이상/일 발생하면 중단 및 원인 분석 후 재개

귀무가설 ($H_0$): 메인 홈 검색 모듈의 위치(하단 vs 최상단)와 상세 페이지 진입 여부 사이에는 상관관계가 없다. 즉, 두 그룹의 상세 페이지 진입률($CVR$)은 동일하다.$$H_0: p_A = p_B$$대립가설 ($H_1$): 메인 홈 검색 모듈을 최상단에 노출하는 것은 상세 페이지 진입 여부에 영향을 미친다. 즉, 실험군(B)의 상세 페이지 진입률이 대조군(A)보다 높을 것이다.$$H_1: p_A < p_B$$

## Q3-1. 위 가설에 적합한 검정 이름과 왜 해당 검정을 써야하는지 이유도 작성해주세요. (t test와 카이제곱검정 중)

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 위 가설에 필요한 검정은 카이제곱(χ²) 검정 이고,  

이유는 모듈 위치(A/B)도 범주형 변수, 상세 페이지 진입 여부(진입/미진입)도 범주형 변수이므로 두 범주형 변수 간 독립성(연관성) / 비율 차이를 보기 위해  분할표(교차표)를 만들고 **진입률이 동일한지(p_A = p_B)를 **검정하는 상황이기 때문 이다.

<br>

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

```python
### A3-4. 코드작성

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (비로그인 유저 기준)
# [카드 클릭(상세진입) 성공 수, 카드 클릭 실패 수]
control_results = [ 159 , 5139 ]   # 대조군(A)
variant_results = [ 212 , 5086 ]   # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 필터 UI 통합(메인 박스 필터화 & 좌측 라디오 삭제)이 카드 클릭률(상세 진입률)에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 필터 UI 통합의 효과가 통계적으로 유의하다고 보기 어렵습니다.")


#############
```
