## A1. 답안 작성  
답변:
A/B 테스트는 변수 A에 비해 대상이 변수 B에 대해 보이는 응답을 테스트하고, 두 변수 중 어떤 것이 더 효과적인지를 판단함으로써 단일 변수에 대한 두 가지 버전을 비교하는 방법이다. 주요 측정지표를 기반으로 변형(B) 버전과 비교하여 컨트롤(A) 버전을 검증하여 최적의 고객 경험을 제공한다.

기존 버전에서 변화한 B 버전을 테스트할 경우, 사용자들이 익숙한 것을 선호하고 변화를 꺼리는 초두효과나 새로운 것에 관심을 가지는 신기효과가 발생하여 실제 영향과 다른 일시적인 트렌드를 생성할 수 있어 충분한 시간과 표본을 가지고 테스트를 진행해야한다.

A/B 테스트는 과거부터 활용되었는데 실제로 2008년 오바마 대통령 선거 운동 중, 기부 웹사이트에서 '가입하기' 버튼의 문구(Sign Up vs. Learn More)와 배경 이미지(가족 사진 vs. 연설 사진)를 테스트하여, 클릭률이 가장 높은 조합을 찾아내 후원금을 증가시키며 효과적으로 A/B 테스트를 활용하였다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:

p1
  (Control Group, 대조군): 기존 UI(좌측 사이드바)에서의 공모전 상세 페이지 진입 전환율(CVR)입니다.

p2
  (Experimental Group, 실험군): 새로운 UI(상단 필터 칩)를 적용했을 때 기대하는 목표 전환율입니다.

가정된 값:

p1
 = 0.15 (15%): 현재 메인 페이지 방문자 중 약 8%가 공모전 상세 페이지를 클릭한다고 가정했습니다.

p2
 = 0.17 (17%): 필터 접근성이 좋아지면 CVR이 2%p 상승할 것이라고 가설을 세웠습니다. (최소 감지 효과, MDE)

```````````python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.
import math
from scipy.stats import norm

p1 = 0.15
p2 = 0.17
alpha = 0.05
beta = 0.2

z_alpha_half = norm.ppf(1 - alpha / 2)


z_beta = norm.ppf(1 - beta)

numerator = (z_alpha_half + z_beta)**2 * (p1 * (1 - p1) + p2 * (1 - p2))

denominator = (p1 - p2)**2

n = numerator / denominator

print(f"필요한 최소 표본 크기(n): 그룹당 약 {math.ceil(n):,}명")











#########
```````````

귀무가설 ($H_0$): 메인 홈 검색 모듈의 위치(하단 vs 최상단)와 상세 페이지 진입 여부 사이에는 상관관계가 없다. 즉, 두 그룹의 상세 페이지 진입률($CVR$)은 동일하다.$$H_0: p_A = p_B$$대립가설 ($H_1$): 메인 홈 검색 모듈을 최상단에 노출하는 것은 상세 페이지 진입 여부에 영향을 미친다. 즉, 실험군(B)의 상세 페이지 진입률이 대조군(A)보다 높을 것이다.$$H_1: p_A < p_B$$

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 (#####) 이고,  

이유는 `````````` 이다.

<br>

```````````python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```````````

```````````python
### A3-4. 코드작성

import pandas as pd
from scipy.stats import chi2_contingency

control_results = [1751, 4020]  # 대조군(A)
variant_results = [1800, 3471]  # 실험군(B)

data = [control_results, variant_results]


chi2, p_value, dof, expected = chi2_contingency(data)

print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 카테고리 필터 칩(Chips) UI 변경이 상세 페이지 진입 전환율(CVR)을 통계적으로 유의미하게 변화시켰습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> UI 변경에 따른 전환율 차이가 통계적으로 증명되지 않았습니다.")









#############
```````````
